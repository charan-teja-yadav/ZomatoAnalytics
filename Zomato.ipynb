```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

Loading the Dataset

```python
df = pd.read_csv('Zomato data .csv')
```

```python
df
```

Handling the Rate

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)
```

```python
df['rate'] = df['rate'].apply(handleRate)
```

```python
df
```

```python
df
```

```python
df[df['rate']< 3]
```

Types of restaurant 

```python
sns.countplot(x = df['listed_in(type)'],palette=['red', 'green', 'blue','yellow'])
warnings.filterwarnings('ignore')

plt.xlabel('types of restaurant')
plt.show()
```

Majority restaurants fall in Dining

```python
grouped = df.groupby('listed_in(type)')['votes'].sum()
```

```python
result = pd.DataFrame({'votes':grouped})
plt.plot(result,c = 'green',marker = "o")
X=plt.xlabel('listed_in(type)',c = 'red',size = 20)
Y=plt.ylabel('votes',c = 'red',size = 20)
plt.title("Count of votes of restaurant type")
```

```python
plt.hist(df['rate'],bins=5)
plt.title('Majority of rating restaurants receive')
plt.show()
```

Majority of the ratings received are between 3.5 and 4.0

```python
couple_data = df['approx_cost(for two people)']

sns.countplot(x = couple_data)
plt.title('Average spending on Orders')
```

```python
df
```

Which mode receives maximum rating

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order',y = 'rate',data = df,palette='Set1')
plt.show()
```

Conclusion: Offline order received lower rating compared to online order

```python
pivot_table = df.pivot_table(index = 'listed_in(type)',columns = 'online_order',aggfunc = 'size',fill_value=0)
sns.heatmap(pivot_table,annot = True,cmap = 'YlGnBu',fmt = 'd')
plt.title('Heatmap')
plt.xlabel('online_order')
plt.ylabel('listed_in(type)')
plt.show()
```

Conclusion: As per the heatmap, Dining restaurants received more offline orders,From this we understood that people prefer to order snacks online from cafes where they are intrested to come in person for the restaurants
